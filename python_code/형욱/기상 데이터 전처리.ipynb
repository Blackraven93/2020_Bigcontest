```python
dust_var_select = dust[['일시', '1시간평균 미세먼지농도(㎍/㎥)']]
dust_var_select
```

- 한글로 된 변수명들을 영어로 변경해주겠다.

# 작업해 준 기상 데이터와 황사(미세먼지) 데이터를 결합

- 결합하기 전에 먼저 두 데이터 셋에 존재하는 date 변수(조인키)의 타입을 확인해봐야 한다.
    - 불일치할 경우, join이 제대로 수행되지 않기 때문에 확인 필수!!

```python
weatehr_var_select.info()
```

```python
dust_var_select.info()
```

- weatehr_var_select 데이터를 기준으로 두 데이터 셋을 **left join** 해 줄 것이다.
    - Left join을 해주는 이유는 weatehr_var_select 데이터 셋의 길이는 **8761개**인데, dust_var_select 데이터 셋의 길이는 **8200개**이다.
    - 따라서 더 많은 데이터가 존재하는 weatehr_var_select 데이터 셋을 기준으로 left join 해주는 것이다.
    - 조인키는 date 변수를 사용하겠다.

```python
weather_final = pd.merge(weatehr_var_select, dust_var_select, how = 'left', on = 'date')
weather_final
```

```python
weather_final.info()
```

- date 변수의 변수 타입을 확인해보면, object라고 되어있음을 확인할 수 있다.
- 그런데 NS Shop+ 매출 실적 데이터 셋의 date 변수의 변수 타입은 datetime이다.
    - 따라서 **date 변수의 변수 타입을 datetime으로 변경**해주도록 하겠다.

```python
weather_final['date'] = pd.to_datetime(weather_final['date'], format = '%Y-%m-%d %H:%M:%S')
weather_final.info()
```

```python
# date 변수의 변수 타입이 datetime으로 잘 변경되었는지 확인
weather_final
```

# 비 내림 여부 및 눈 내림 여부 변수 생성
- 현재 데이터 셋에 존재하는 rain, snow 변수에는 결측값들이 굉장히 많다.
- 따라서 **해당 일에 비 또는 눈이 한 시간대라도 내렸으면 "1", 그렇지 않으면 "0"**으로 처리해주는 **binary 변수**를 생성해서 사용하는 것이 추후에 모델의 예측 성능을 높이는 데 도움이 될 것으로 판단된다.
    - **비 내림 여부(is_rain)**
        - 0: 비 내리지 않음
        - 1: 비 내림
    - **눈 내림 여부(is_snow)**
        - 0: 눈 내리지 않음
        - 1: 눈 내림

- 앞서 말했듯이, rain 및 snow 변수에 굉장히 많은 결측값들이 존재한다.

```python
weather_final.isnull().sum()
```

- 먼저 date 변수를 활용하여 year, month, day 변수를 생성해주도록 하겠다.
    - 년/월/일 변수를 결합한 뒤(new_date), 새로 생성된 new_date 변수를 기준으로 비 내림 및 눈 내림 여부 변수를 만들어 주도록 하겠다.

```python
weather_final['year'] = weather_final['date'].dt.year
weather_final['month'] = weather_final['date'].dt.month
weather_final['day'] = weather_final['date'].dt.day

weather_final
```

```python
# 아래의 for문을 돌리기 위해 임시적으로 "0" 값으로 채워진 new_date 변수를 미리 만들어 주었음
weather_final['new_date'] = 0

for i in tqdm(range(len(weather_final))):
    weather_final['new_date'][i] = str(weather_final['year'][i]) + '-' + str(weather_final['month'][i]) + '-' + str(weather_final['day'][i])
weather_final
```

- 편리성을 위해 rain 및 snow 변수의 NaN 값들을 모두 "0"으로 바꿔준 뒤 작업을 시작하도록 하겠다.
    - 단, 기존의 rain 및 snow 변수는 건드리지 않고 rain_fillna, snow_fillna 변수를 새롭게 만들어서 사용하도록 하겠다.

```python
weather_final['rain_fillna'] = weather_final['rain'].fillna(0)
weather_final['snow_fillna'] = weather_final['snow'].fillna(0)
weather_final.isnull().sum()
```

```python
# 시간대별(date)로 존재했던 강수량 및 적설량을 일자별(new_date)로 합쳐주기(sum) 
weather_pivot = pd.pivot_table(weather_final, index = 'new_date', aggfunc = np.sum)
# 비 내림 및 눈 내림 여부 생성을 위해 필요한 변수들만 선택 
weather_pivot = weather_pivot[['rain_fillna', 'snow_fillna']]
weather_pivot
```

```python
# 아래의 for문을 돌리기 위해 임시적으로 "0" 값으로 채워진 is_rain, is_snow 변수를 미리 만들어 주었음
weather_pivot['is_rain'] = 0
weather_pivot['is_snow'] = 0

for i in range(len(weather_pivot)):
    # 비가 내리지 않은 경우
    if weather_pivot['rain_fillna'][i] == 0:
        weather_pivot['is_rain'][i] = 0
    # 비가 내린 경우
    else:
        weather_pivot['is_rain'][i] = 1
        
    # 눈이 내리지 않은 경우
    if weather_pivot['snow_fillna'][i] == 0:
        weather_pivot['is_snow'][i] = 0
    # 눈이 내린 경우
    else:
        weather_pivot['is_snow'][i] = 1
```

```python
weather_pivot['is_rain'].value_counts()
```

```python
weather_pivot['is_snow'].value_counts()
```

```python
df_is_rain_snow = weather_pivot.reset_index()
df_is_rain_snow = df_is_rain_snow[['new_date', 'is_rain', 'is_snow']]
df_is_rain_snow
```

- 이제 만들어진 df_is_rain_snow 데이터 셋과 기존의 weather_final 데이터 셋을 결합해주도록 하겠다.
    - new_date 변수를 조인키로 사용하여 **inner join**을 수행해주도록 하겠다.

```python
weather_final2 = pd.merge(weather_final, df_is_rain_snow, how = 'inner', on = 'new_date')
weather_final2
```

- 위 결과를 보면 알 수 있듯이, inner join이 잘 수행되었다.
- 이제 is_rain, is_snow 변수 생성을 위해 **임시로 만들어 놓은 "year, month, day, new_date, rain_fillna, snow_fillna" 변수들은 제거**해주도록 하겠다.

```python
weather_final2.drop(['year', 'month', 'day', 'new_date', 'rain_fillna', 'snow_fillna'], axis = 1, inplace = True)
weather_final2
```
